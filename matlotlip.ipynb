```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data=[11,14,33,44,55,67,11,23,34,65,64,44,44,65,56,12,23,43,69,1,2,34,45]
plot=['histogram','KERNAL_DENSITY']
for plots in plot:
    if plots =='Histogram':
        sns.histplot(data)
        plt.title("Histogram chart")
        plt.show()
    elif plots =='KERNAL_DENSITY':
        sns.kdeplot(data)
        plt.title("KERNAL_DENSITY")
        plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Names
names = ['Aarav','Vivaan','Aditya','Arjun','Krishna','Ishaan','Shaurya','Atharv','Dhruv','Kabir',
         'Rohan','Kunal','Rahul','Ankit','Saurabh','Vikas','Aman','Deepak','Nikhil','Sumit',
         'Pankaj','Ravi','Manish','Yash','Harsh']

# Salary (5000 - 50000 range)
salary = np.array([12000,25000,18000,32000,27000,15000,40000,22000,35000,29000,
                   31000,17000,26000,45000,38000,21000,14000,33000,28000,36000,
                   19000,24000,50000,23000,34000])

# Create DataFrame
df = pd.DataFrame({
    'Name': names,
    'Salary': salary
})

print(df)

# ---------------- BAR CHART ----------------
plt.figure(figsize=(10,5))
plt.bar(df['Name'], df['Salary'])
plt.xticks(rotation=90)
plt.title("Salary of Employees")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.show()

# ---------------- PIE CHART ----------------
plt.figure(figsize=(8,8))
plt.pie(df['Salary'], labels=df['Name'], autopct='%1.1f%%')
plt.title("Salary Distribution")
plt.show()

# ---------------- SCATTER PLOT ----------------
plt.figure(figsize=(8,5))
plt.scatter(range(len(df)), df['Salary'])
plt.title("Scatter Plot of Salary")
plt.xlabel("Index")
plt.ylabel("Salary")
plt.show()
```
