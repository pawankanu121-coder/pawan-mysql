```python
import matplotlib.pyplot as plt  #pyplot is a collection that make matplotlib work as MATLAB
```

```python
#data
x=[1,2,3,4,5]
y=[10,20,30,35,44]

plt.plot(x,y) #line chart
plt.show() #display chart in jp nb
```

```python
plt.figure(figsize=(4,3))  #figure size
plt.plot(x,y,color='blue',marker='o',linestyle='--',linewidth=2,markersize=12)

plt.title("kishan")
plt.xlabel("x-axis label hai")
plt.ylabel("y-axis label hai")

plt.show()
```

```python
#data
x=[1,2,3,4,5]
y=[10,20,30,35,44]
y2=[20,30,35,45,55]

plt.plot(x,y,label='sales 2024') #ploting y data
plt.plot(x,y2,label='sales 20252') #ploting y2 data

plt.title("yoy sales")
plt.xlabel("months")
plt.ylabel("sales")
plt.legend() #show legends
plt.show()
```

```python
#use for distribution analysis
#data
import random
data =[random.randint(1, 20) for _ in range(500)]

plt.hist(data,bins=20) #histagram chart
plt.title("Histogram Example")
plt.show()
```

```python
#used to show paart -to - whole relationships/data

#data
categories=['A','B','C','D','E']
sales=[10,20,55,35,45]
plt.pie(sales,labels=categories,autopct='%1.1f%%')
plt.title("Sales Pie Chart")
plt.show()
```

```python
#used to find relationship b/w  varibales
#data
y1=[10,20,30,40,50]
y2=[20,30,35,45,55]
plt.scatter(y1,y2)
plt.title('scalter plot Example')
plt.show()
```

```python
#used to show multiple charts in one figure
#data-1 - bar chart
categories=['mon','tue','wed','thur','fri']
sales=[10,20,55,35,45]


#data- 2- scatter plot
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]

plt.figure(figsize=(10,4))

#first plot-bar chart
plt.subplot(1,2,1)  #rows, column,position
plt.bar(categories,sales)
plt.title('daily sales')
plt.xlabel('week days')
plt.ylabel('sales')

#second plot-scatter chart
plt.subplot(1,2,2) #row, column, position
plt.scatter(y1,y2)
plt.title('user Example')
plt.xlabel('uuser1')
plt.ylabel('user2')

plt.tight_layout()
plt.show()
```
